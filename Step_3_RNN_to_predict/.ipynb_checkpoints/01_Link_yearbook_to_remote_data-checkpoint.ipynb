```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the remote-sensing and yearbook data
remote_data = pd.read_csv('../Step_2_Calculate_MODIS_with_admin_boundary/Data/without_cloud_mask/RS_data_selected.csv',parse_dates=['Date'])
yrbook_data = pd.read_csv('../Step_2_Calculate_MODIS_with_admin_boundary/Data/without_cloud_mask/Yrbook_data.csv')
```

##### Link Yearbook to remote_data

```python
# filter the unit yield of Wheat and Corn 
mean_yield = yrbook_data[yrbook_data['Metrics'] == 'Unit yield (Kg/ha)']
```

```python
# create a new colum for holding the input time series features
mean_yield['In_feature'] = np.nan


# define the lenth of growth days for wheat and corn
feature_len_wheat = 165 # Jan-01 to June-15
feature_len_corn  = 108  # Jun-15 to Oct-01


# loop through each year
for year in yrbook_data['Year'].unique():
    # loop through each crop
    for crop in yrbook_data['Crop'].unique(): 
        
        # get the yr_data with spicific year-crop
        yr_crop = mean_yield[(mean_yield['Year']==year)&
                             (mean_yield['Crop']==crop)]        
        # get all county names given the year-crop
        county_names = yr_crop['Region'].unique()

        # loop through each county
        for name in county_names:
            # get the feature with specific year-name
            all_feature = remote_data[(remote_data['Date'].dt.year == year)&
                                      (remote_data['NAME'] == name)]
            
            # condition to test if there are remote sensed date with given year-name
            if len(all_feature)>0:
                # get the infeatures for wheat and corn
                in_feature_wheat = all_feature[(all_feature['Date'] <f'{year}-06-15')][-feature_len_wheat:]
                in_feature_corn  = all_feature[(all_feature['Date'] >=f'{year}-06-15')][:feature_len_corn]
                # insert the remote_data to yr_book_data
                idx = mean_yield.loc[(mean_yield['Year']==year)&
                                     (mean_yield['Region']==name)&
                                     (mean_yield['Crop']==crop)].index[0]
                
                # judge the crop type
                if crop == 'Wheat': 
                    mean_yield.loc[idx,'In_feature'] = [in_feature_wheat]
                elif crop == 'Corn': 
                    mean_yield.loc[idx,'In_feature'] = [in_feature_corn]
                    
                # print out the process
                print(f'{year}_{name}_{crop}')
            else:
                print(f'{year}_{name}_{crop} <<<<<<<<<<<<<<<<<<No matching remote sensed data>>>>>>>>>>>>>>>>')
```

```python
# save the df to pickle, so the data format can be reserverd
mean_yield.to_pickle('./Data/Yearbook_link_remote.pkl')

# read pickel
mean_yield_yrbook = pd.read_pickle('./Data/Yearbook_link_remote.pkl')
mean_yield_yrbook.reset_index(inplace=True,drop=True)
```
